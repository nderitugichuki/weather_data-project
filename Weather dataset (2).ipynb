DATA VISUALIZATION
#Histplot

```python
#distribution of temperature
plt.figure(figsize=(10,6))
sns.histplot(df['Temp_C'], bins=20,kde=True, color='skyblue')
plt.title('Distribution of Temperature (C)')
plt.xlabel('Temeprature (C)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Distribution for humidity
plt.figure(figsize=(10,6))
sns.histplot(df['Rel Hum_%'], bins=20,kde=True,color='lightcoral')
plt.title('Distribution of Relative humidity (%)')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Distribution for wind speed
plt.figure(figsize=(10,6))
sns.histplot(df['Wind Speed_km/h'], bins=20,kde=True,color='lightgreen')
plt.title('Distribution of Wind Speed (km/h)')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grind=True()
plt.show()
```

```python
#Distribution for wind speed
plt.figure(figsize=(10,6))
sns.scaterplot(df['Wind Speed_km/h'],color='lightgreen')
plt.title('Distribution of Wind Speed (km/h)')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.grind=True()
plt.show()
```

Time series graph

```python
#set date as the index
df.set_index('Date/Time', inplace=True)
```

```python
#Time series plot for tempearature
plt.figure(figsize=(14,7))
plt.plot(df.index,df['Temp_C'], color='tab:blue')
plt.title('Time series for temperature (C)')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (C)')
plt.grid=True
plt.show()
```

```python
#time series plot for humidity
plt.figure(figsize=(14,7))
plt.plot(df.index,df['Rel Hum_%'], label='Relative Humidity (%)', color='tab:orange')
plt.title('Time series of Relative humidity (%)')
plt.xlabel('Date/Time')
plt.ylabel('Rel Hum_%')
plt.grid=True
plt.show()
```

```python
#time series plot for wind speed
plt.figure(figsize=(14,7))
plt.plot(df.index,df['Wind Speed_km/h'], label='Wind Speed km/h',color='tab:green')
plt.title('Time series for Wind Speed (km/h')
plt.xlabel=('Date/Time')
plt.ylabel=('Wind Speed (km/h)')
plt.grind=True
plt.show()
```

CORRELATION MATRIX


```python
#select numerical columns
numerical_df=df.select_dtypes(include=['number'])
correlation_matrix=numerical_df.corr()
print(correlation_matrix)
```

```python
#Heatmap plot
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt= '.2f')
plt.title('Correlation Matrix for numerical columns')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
