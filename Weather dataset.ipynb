```python
#renaming coulumns
df=df.rename(columns={'Weather':'Weather_Condition'})
```

```python
#calculate the mean visibility
mean_visibility=df['Visibility_km'].mean()
print(mean_visibility)
```

```python
#Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km
filtered_df=df[(df['Wind Speed_km/h']>24)&(df['Visibility_km']==25)]
print(filtered_df)
```

```python
#number of records
record_count=filtered_df.shape[0]
print(record_count)
```

```python
#mean value of each column for each weather condition
mean_values=df.groupby('Weather_Condition').mean(numeric_only=True)
print(mean_values)
```

```python
#Instances where the weather is clear and the relative humidity is greater than 50.
condition_1=(df['Weather_Condition']=='Mainly Clear') & (df['Rel Hum_%']>50)
print(condition_1)
```

```python
#visibility is above 40
condition_2=(df['Visibility_km']>40)
print(condition_2)
```

```python
#Applying the final condition to filter the dataframe
final_condition = condition_1 | condition_2
print(final_condition)
```

```python
#filtered dataframe
df2=final_condition
df[final_condition]
```

```python
#Total instances satisfying both conditions
instances=final_condition.sum()
print(instances)
```

```python
#Number of weather conditions that include snow
Snow_conditions=df['Weather_Condition'].str.contains('Snow')
print(Snow_conditions)
```
